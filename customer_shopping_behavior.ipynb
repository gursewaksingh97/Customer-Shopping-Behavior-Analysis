```python
# df['Review Rating']=df['Review Rating'].fillna(df['Review Rating'].mode()[0]
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# colms name changing
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
#  create age groups 
labels=['Young Adult','Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].sample(10)
```

```python
df.sample(4)
```

```python
# create colmn purchase_frequency_days 
freq_mapping={
    'Weekly': 7,
    'Bi-Weekly': 14,
    'Fortnightly': 14,
    'Monthly': 30,
    'Every 3 Months':90,
    'Quarterly': 90,
    'Annually':365
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(freq_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].sample(5)
```
